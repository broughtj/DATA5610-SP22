```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [10, 8]
```

```python
%matplotlib inline
```

## Simulate Two Totally Independent RW Processes

```python
n = 252
m = 10_000

betas = np.zeros(m)
rsqrs = np.zeros(m)

for i in range(m):
    y = np.cumsum(np.random.normal(size=n))
    x = np.cumsum(np.random.normal(size=n))
    results = sm.OLS(y, x).fit()
    betas[i] = results.params[0]
    rsqrs[i] = results.rsquared
```

```python
bs = pd.Series(betas)
bs.plot.kde()
```

```python
rs = pd.Series(rsqrs)
rs.plot.kde()
```
